```python
# Set python path here
import sys
sys.path.append('/home/narihira/workspace/caffe-bvlc/python')
import matplotlib.pyplot as plt
import cv2
import numpy as np
import caffe
import direct_intrinsics as di
```

```python
# Load net
modelfile = 'mit_final_barron_train.caffemodel'
net = caffe.Net('direct_intrinsics.prototxt', modelfile, caffe.TEST)
```

```python
# Load an image
obj_id = 7
obj = di.mit_objects['test'][obj_id]
print obj
img = cv2.imread('/home/narihira/data/DI-MIT-intrinsic/data/{}/diffuse.png'.format(obj))
mask = cv2.imread('/home/narihira/data/DI-MIT-intrinsic/data/{}/mask.png'.format(obj))
mask[mask > 0] = 1
```

```python
# Predict by Direct intrinsics CNN
a, s = di.predict(net, img)
```

```python
# Display results
plt.figure()
plt.title('input')
plt.imshow(di.bgr2rgb(img))
plt.figure()
plt.title('albedo')
plt.imshow(mask * a)
plt.figure()
plt.title('shading')
plt.imshow(mask * s)
```
